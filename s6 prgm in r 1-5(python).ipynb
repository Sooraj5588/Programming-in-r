Exp 4: Pascals triangle

```python
n=5
for i in range(0,n+1):
    print(" "*(n-i), end="")
    for j in range(1,i+1):
        print(j,end=" ")
    print()
```

Exp 6: Palindrom and Factorial

```python
a=input("Enter the Sentence")
b=len(a)
c=0
for i in range(0,b):
    if a[i]==a[b-i-1]:
        c=c+1
    else:
        print("Not Palindrome")
        break
if c==b :
    print("Palindrome")
```

```python
n=int(input("Enter the number"))
fact=1
for i in range (1,n+1):
    fact=fact*i
print("The factorial of ",n,"is",fact)
```
